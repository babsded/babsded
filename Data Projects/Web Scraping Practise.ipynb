```python
def strip_newline(text):
    if type(text) is str:
        return text.strip()
    else:
        pass
```

```python
df['Job Title'] = df['Job Title'].apply(lambda row: row.strip())
#OR 
df['Job Title'] = df['Job Title'].apply(strip_newline)
```

```python
df['Company'] = df['Company'].apply(strip_newline)
df['Applicants'] = df['Applicants'].apply(strip_newline)
df['Recruiter Statistics'] = df['Recruiter Statistics'].apply(strip_newline)
df['Location'] = df['Location'].apply(strip_newline)
df['Published'] = df['Published'].apply(strip_newline)
df['Other Job Openings at the Company'] = df['Other Job Openings at the Company'].apply(strip_newline)
```

```python
for col in df.columns:
    df[col] = df[col].apply(strip_newline)
```

```python
s = ['Salary', 'Published','Work Day', 'Employment type', 'Positions available','Minimal experience', 'Gender']
for col in s:
    df[col] = df[col].apply(strip_newline)
```

```python
df['Applications'] = df['Applicants'].apply(lambda row: row.split(' | ')[0].split(': ')[1] if row is not None else None)
df['Visits'] = df['Applicants'].apply(lambda row: row.split(' | ')[1].split(': ')[1] if row is not None else None)
```

```python
df['CV seen'] = df['Recruiter Statistics'].apply(lambda row: row.split(',')[0].split(': ')[1] if row is not None else None)
df['CV deleted'] = df['Recruiter Statistics'].apply(lambda row: row.split(',')[1].split(': ')[1] if row is not None else None)
df['CV unseen'] = df['Recruiter Statistics'].apply(lambda row: row.split(',')[2].split(': ')[1] if row is not None else None)
```

```python
def get_min_age(text):
    if text is None:
        return None
    else:
        if len(text.split(' to ')[0].split(' ')) == 1:
            return 17
        else:
            return text.split(' to ')[0].split(' ')[1]
```

```python
def get_max_age(text):
    if text is None:
        return None
    else:
        return text.split(' to ')[1].split(' ')[0]
```
